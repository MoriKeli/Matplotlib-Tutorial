### Labeling x & y axes.
#### In lesson 1, we displayed a graph without labels and a title. If you were to present the graph to someone, he will be confused what the graph is about (e.g. Does the graph rep. population of a certain town or GDP of a certain country or mortality rate, etc. ???)

#### To make your graphs more readable and understable, you need to give it a title and label the x and y axes.

```python
# Learning Matplot using Jupyter Notebook
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# graph labels and values
labels, values = ['A', 'B', 'C'], [10, 20, 30]
```

#### To label our graph, we use the following Matplotlib functions:

  #### plt.title() to give it a title
  #### plt.xlabel() to label the x-axis
  #### plt.ylabel() to label the y-axis
    

```python
# plotting basic graph
plt.title('Basic Graph')
plt.xlabel('Values')
plt.ylabel('Labels')

plt.plot(labels, values)
plt.show()
```
